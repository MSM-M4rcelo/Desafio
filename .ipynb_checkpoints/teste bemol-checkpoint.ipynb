# Questão 1

```python
import pandas as pd
```

```python
vendas_df = pd.read_excel('desafio.xlsx')
display(vendas_df)
```

```python
print(vendas_df.info())
```

# Percentual de Imposto

```python
vendas_df['Imposto'] = vendas_df.apply(lambda x: (x['Imposto'] / x['Valor de Compra']) * 100, axis = 1)
display(vendas_df)
```

```python
Impostos= vendas_df['Imposto']
display(Impostos)
#percentual de imposto é 4,5%
```

# Produto mais Vendido

```python
vendas_df['Tipo de Mercadoria'].value_counts()

# os produtos mais vendidos foram SPLIT 18001 À 30000 e SPLIT 10001 À 18000
```

```python
import plotly.express as px
coluna = "Tipo de Mercadoria"
grafico = px.histogram(vendas_df, x=coluna)
grafico.show()
```

# Percentual de CPF na Nota

```python
display(vendas_df["CPF na Nota?"].value_counts())
display(vendas_df["CPF na Nota?"].value_counts(normalize=True).map("{:.1%}".format))
```

# Percentual de Produtos Devolvidos

```python
display(vendas_df["Produto Devolvido"].value_counts())
display(vendas_df["Produto Devolvido"].value_counts(normalize=True).map("{:.1%}".format))
```

# Vendas por vendedor

```python
display(vendas_df["Nome"].value_counts())
display(vendas_df["Nome"].value_counts(normalize=True).map("{:.1%}".format))
```

# Filial que Mais Vendeu

```python
display(vendas_df["Filial"].value_counts())
display(vendas_df["Filial"].value_counts(normalize=True).map("{:.1%}".format))
```
